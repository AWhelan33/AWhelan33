```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_scheduled_monuments_in_Blaenau_Gwent'
```

```python
def scrapeDataAndCreateDataFrame(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'class': "wikitable"})
    df = pd.read_html(str(table))
    df = pd.DataFrame(df[0])
    return df
```

```python
from geopy.geocoders import Nominatim
import numpy as np
```

```python
geolocator = Nominatim(user_agent='app')
```

```python
lat=[]
lon=[]
for community in df['Community']:
    result = geolocator.geocode(community, country_codes="gb")
    if result is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(result.latitude)
        lon.append(result.longitude)
```

```python
print(lat, lon)
```

```python
df['lat']=lat
df['lon']=lon
```

```python
df.to_csv('monuments_list.csv', index=False)
```
